Write a program that calculates the area of a circle based on the radius entered by the user.

```python
def area():
    r = int(input("Enter your radius : "))
    A = r*r*np.pi
    return A
```

Write a program to count the number 4 in a given list.

```python
def count_4():
    l = input("Enter a list of numbers separated by spaces: ")
    l1 = []
    for i in l.split():
        if int(i) == 4:
            l1.append(i)
    return len(l1)
```

```python
count_4()
```

Write a program that computes the greatest common divisor (GCD) and least common multiple (LCM) of two positive integers.

```python
def divisors(a):
    divisors = []
    for i in range(1,a+1):
        if a%i == 0:
            divisors.append(i)
    return divisors
```

```python
divisors(24)
```

```python
def gcd_lcm():
    a = int(input("Enter your first number : "))
    b = int(input("Enter your second number : "))
    l1 = divisors(a)
    l2 = divisors(b)
    l3 = []
    for i in l1:
        for j in l2:
            if i == j:
                l3.append(i)
    c = str(l3[-1])            
    d = str(int((a*b)/l3[-1]))
    print("gcd : "+c, "lcm : "+d)
```

```python
gcd_lcm()
```

Write a program to calculate sum of digits of a number.

```python
def sum_of_digits():
    a = int(input("Enter your number : "))
    n = 0
    while a != 0:
        n += (a%10)
        a = a//10
    return n    
```
