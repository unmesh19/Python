```python
# Perfect number
```

```python
def is_perfect(k):
    if k > 1:
        divisor = []
        for i in range(1,k-1):
            if k % i == 0:
                divisor.append(i)
        if sum(divisor) == int(k):
            return print(k, "is a perfect number")
        else:
            return print(k, "is not a perfect number")
    else:
        return print(k, "is not a perfect number")
```

```python
is_perfect(28)
```

```python
is_perfect(1)
```

```python
is_perfect(12)
```

```python
def armstrong_num(d):
    digits = []
    c = d
    while c > 0:
        digit = c % 10
        digits.append(digit)
        c //= 10
    digit = list(reversed(digits))
    a = sum([digits[i]**len(digits) for i in range(len(digits))])
    if a == d:
        print(d, "is an Armstrong number")
    else:
        print(d, "is not an Armstrong number")
```

```python
armstrong_num(124)
```

```python
armstrong_num(370)
```
