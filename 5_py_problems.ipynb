```python
# Create a list of squared integers from a given list of integers.
```

```python
def Sqr_list(l):
    sqr_list = []
    for i in l:
        sqr_list.append(i**2)
    return sqr_list
```

```python
l = [1,4,7,9]
Sqr_list(l)
```

```python
# Create a list of vowels from given string.
```

```python
def vowels(l):
    l = l.lower()
    vowels = []
    for i in range(len(l)):
        char = l[i]
        if char in "aeiou":
            vowels.append(char)
    return print(vowels)
```

```python
a = "Hello word"
vowels(a)
```

```python
# Create a list of all the positive integer from a given list of integers.
```

```python
def positive_integer(l):
    positive_list = []
    for i in l:
        if i >= 0:
            positive_list.append(i)
    return positive_list
```

```python
l = [-4,-3,-2,-1,0,1,2,3,4]
positive_integer(l)
```

```python
# Create a list of all the unique words from a given sentence.
```

```python
def unique_words(l):
    a = l.split(" ")
    unique = set(a)
    return unique
```

```python
a = "Hello Word I Am Unmesh Hardikar"
unique_words(a)
```
