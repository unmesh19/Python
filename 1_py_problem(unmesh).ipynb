```python
# Given an array arr[] of N integers, calculate the median
```

```python
def median(l):
    if len(l)%2 == 0:
        a = int(len(l)/2) - 1
        return (l[a]+l[a+1])/2 
    else:
        b = len(l)//2
        return l[b]   
```

```python
l = [12, 23, 46, 56, 67]
median(l)
```

```python
l1 = [125, 233, 934, 454, 556, 687]
median(l1)
```

```python
# Given an integer x, return true if x is a palindrome, and false otherwise.
```

```python
def is_palindrome(x):
    if int(str(x)[::-1]) == x:
        return print(x, "is a palindrome.")
    else:
        return print(x, "is not a palindrome.")
```

```python
is_palindrome(54645)
```

```python
is_palindrome(12345)
```

```python
# Given a sorted array of size N and an integer K, find the first position at which K is present in the array using binary search
```

```python
def binary_search(l, a):
    l1 = []
    for i in range(len(l)):
        if l[i] == a:
            l1.append(i)
    if len(l1) == 0:
        print(a, "is not in this list")
        return None
    else:
        return l1
```

```python
l = [6,2,0,4,5,8,4,7]
binary_search(l,4)
```

```python
l = [3,4,1,8,9,8]
binary_search(l,6)
```

```python
# Given a number and its reverse. Find that number raised to the power of its own reverse
```

```python
def num_raised_rev(a):
    b = int(str(a)[::-1])
    return print(a**b)
```

```python
num_raised_rev(10)
num_raised_rev(3)
```
