```python
# Reverse a given string
```

```python
def reversed_list(a):
    b = a.split(" ")
    return b[::-1]
```

```python
a = "Hello Word I Am Unmesh Hardikar"
reversed_list(a)
```

```python
# Create a list of all the even integer from a given list of integers.
```

```python
def Even_list(l):
    even_list = []
    for i in l:
        if i%2 == 0:
            even_list.append(i)
    return even_list
```

```python
l = [1,2,3,4,5,6,7,8,9]
Even_list(l)
```

```python
# Create a list of all the uppercase characters from a given string.
```

```python
def is_upper(a):
    l = []
    for i in a:
        if i.isupper():
            l.append(i)
    return l
```

```python
a = "Hello Word I Am Unmesh Hardikar"
is_upper(a)
```

```python
# Create a list of all the first letters of words in a given sentence.
```

```python
def first_letter(a):
    a1 = words(a)
    l = []
    for i in a1:
        b = i[0]
        l.append(b)
    return l  
```

```python
a = "Hello Word"
first_letter(a)
```

```python
# Create a list of all the anagrams of a given word from a list of words.
```

```python
def get_anagrams(word, word_list):
    sorted_word = sorted(word)
    anagrams = []
    for i in word_list:
        if sorted(i) == sorted(word):
            anagrams.append(i)  
    return anagrams     
```

```python
l = ["enlist", "silent", "tinsel", "apple", "insect"]
get_anagrams("listen", l)
```
