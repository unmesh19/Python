```python
# Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
```

```python
def two_sum(A, a):
    for i in A:
        b = a - i
        if b in A:
            return (i,b)
```

```python
nums = [2, 7, 11, 15]
target = 9
two_sum(nums, target)
```

```python
# Given a string s consisting of words and spaces, return the length of the last word in the string.
```

```python
def len_last_word(a):
    b = a.split()
    return len(b[-1])
```

```python
# Given a integer, return a string containing all the digits of given integer
```

```python
def int_to_list(a):
    digits = []
    if type(a) == str:
        for i in a:
            digits.append(i)
    else:
        for i in str(a):
            digits.append(i)
    return digits
```

```python
int_to_list(1101)
```

```python
# Given a binary number in string form, return the number in decimal form.
```

```python
def dec_to_bi(a):
    b = int(a)
    if b == 0:
        return "0"
    digits = []
    while b > 0:
        r = b%2
        digits.insert(0, str(r))
        b //= 2
    output = ''.join(digits)
    return output
```

```python
dec_to_bi("124")
```

```python
# Given a binary number in string form, return the number in decimal form.
```

```python
def bi_to_dec(a):
    A = int_to_list(a)
    B = A[::-1]
    c = 0
    for i in range(len(A)):
        c += ((int(B[i]))*(2**i))
    return c           
```

```python
bi_to_dec("11010")
```

```python
# Given two binary strings a and b, return their sum as a binary string.
```

```python
def add_bi(x, y):
    return dec_to_bi(int(bi_to_dec(x)+bi_to_dec(y)))
```

```python
x = "1101"
y = "0010"
add_bi(x, y)
```
